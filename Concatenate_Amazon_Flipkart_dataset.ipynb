```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
pd.set_option('display.max_colwidth',None)
```

```python
# amazon dataset

dfa = pd.read_csv('Amazon_microwave_oven.csv')
dfa.head()
```

```python
# flipkart dataset

dff = pd.read_csv('Flipkart_microwave_oven.csv')
dff.head()
```

```python
dff.rename(columns={'discounts in %':'discount_percent'},inplace=True)
```

```python
dff.rename(columns={'Model Name':'Model'},inplace=True)
dff.rename(columns={'Type':'Oven Cooking Mode'},inplace=True)
```

```python
dff.rename(columns={'sales_price in Rs/-':'sales_price',
                   'mrp in Rs/-':'mrp',
                   'star_rating of product out of 5':'star_rating'},inplace=True)
```

```python
dff.rename(columns={'Capacity in litres':'Capacity'},inplace=True)
```

```python
# concatenating the two dataframes

df = pd.concat([dff,dfa],axis=0).reset_index(drop=True)
df.head()
```

```python
df.shape
```

```python
df.rename(columns={'Brand':'brand'},inplace=True)
df.rename(columns={'Model':'model_name'},inplace=True)
df.rename(columns={'Oven Cooking Mode':'type_of_oven'},inplace=True)
df.rename(columns={'Capacity':'capacity_litres'},inplace=True)
```

```python
columns_to_keep = ['product_title','product_url','brand','sales_price','mrp','discount_percent','number_of_ratings','number_of_reviews','type_of_oven','capacity_litres','model_name','star_rating']
df_sel = df[columns_to_keep]
```

```python
df_sel.head()
```

```python
df_sel.dtypes
```

```python
df_sel[['sales_price','mrp','discount_percent','capacity_litres','number_of_ratings','number_of_reviews']] = df_sel[['sales_price','mrp','discount_percent','capacity_litres','number_of_ratings','number_of_reviews']].astype('Int64')
```

```python
df_sel.dtypes
```

```python
len(df_sel)
```

```python
df_sel.shape
```

```python
df_sel.head()
```

```python
# Check for duplicates

df_sel.duplicated(subset='product_title').sum()
```

```python
# duplicated rows

df[df_sel.duplicated(subset='product_title')]
```

```python
# drop duplicates

df_sel.drop_duplicates(subset=['product_title'],keep='first',inplace=True)
```

```python
# Calculate the threshold for 70% NaN values
threshold = df_sel.shape[1] * 0.7

# Select rows where more than 70% values are NaN
nan_rows = df_sel[df_sel.isnull().sum(axis=1) >= threshold]
nan_rows
```

```python
len(nan_rows)
```

```python
dfc = df_sel.copy()
```

```python
# dropping the rows with greater than or equal to 70% nan values 

dfc = dfc.drop(nan_rows.index)
```

```python
# final dataset

dfc
```

```python
dfc.to_csv('microwave_oven.csv',index=False)
```
